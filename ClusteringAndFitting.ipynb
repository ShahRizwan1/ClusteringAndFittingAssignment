```python
#Importing Necessory Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression


#Loading data from dataset
DataSet = pd.read_csv('onlinefoods.csv')

#Converting columns to actual datatypes
DataSet['Age'] = DataSet['Age'].astype(int)
DataSet['Family size'] = DataSet['Family size'].astype(int)
DataSet['latitude'] = DataSet['latitude'].astype(float)
DataSet['longitude'] = DataSet['longitude'].astype(float)
DataSet['Pin code'] = DataSet['Pin code'].astype(int)

#Plotting a scatter Plot
sns.set_style("whitegrid")
plot.figure(figsize=(8, 6))
sns.scatterplot(x="Age", y = "Monthly Income", hue="Gender", data=DataSet)
plot.xlabel('Age')
plot.ylabel('Monthly Income')
plot.title('Age vs Monthly Income By Gender')
plot.show()

#Converting catagorical variables to numerical values
encoder = {}
for cols in ['Gender', 'Marital Status', 'Occupation', 'Monthly Income', 'Educational Qualifications', 'Output', 'Feedback']:
    encoder[cols] = LabelEncoder()
    DataSet[cols]=encoder[cols].fit_transform(DataSet[cols])

#Splitting data into featur X and target Y
x = DataSet[['Gender','Marital Status','Occupation','Monthly Income', 'Educational Qualifications', 'Output']]
y = DataSet['Feedback']

#Now train logistic regression classifier
model = LogisticRegression()
model.fit(x,y)

#Prediction on test set
yPredict = model.predict(x)
cm = confusion_matrix(y, yPredict)

#Plotting Pie chart
plot.figure(figsize=(6,6))
labels = ['Postive Feedback', 'Negative Feedback']
sizes = DataSet['Feedback'].value_counts()
plot.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=140)
plot.title('Pie Chart of Feedback')
plot.axis('equal')
plot.show()

#Plotting Confusion Matrix
plot.figure(figsize=(6,6))
sns.heatmap(cm,annot=True, fmt="d", cmap="Blues")
plot.xlabel('Predicted Feedback')
plot.ylabel('True Feedback')
plot.title('Output vs Feedback')
plot.show()

#Plotting Corelation Matrix
plot.figure(figsize=(6,6))
sns.heatmap(DataSet[['Age', 'Family size', 'Monthly Income']].corr(), annot=True, cmap="coolwarm")
plot.title('Correlation Matrix of Age, Family Size and Monthly Income')
plot.show()

#Plotting Histogram 
plot.figure(figsize=(6,6))
sns.histplot(DataSet['Age'], bins=20, kde=True)
plot.xlabel('Age')
plot.ylabel('Frequency')
plot.title('Histogram of Age')
plot.show()


#Now applying k_Mean clustering and fitting technique

X = DataSet[['Age', 'Monthly Income']].values

K_Means = KMeans(n_clusters=2, random_state =42)
K_Means.fit(X)

XK_Means = K_Means.transform(X)
plot.figure(figsize=(8,6))
plot.scatter(X[:, 0], X[:, 1], c=K_Means.labels_, cmap="coolwarm")
plot.title("K-Mean_Clusters and Cluster_Centers")
plot.xlabel('Age')
plot.ylabel('Monthly Income')

#Applying Fitting Tech
x = np.array(X[:, 0].reshape(-1, 1))
y = np.array(X[:, 1].reshape(-1, 1))

linearRegres = LinearRegression()
linearRegres.fit(x, y)
plot.plot(x, linearRegres.predict(x), color = 'red')
plot.show()


```
